```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


match_data = pd.read_csv('../data/preprocessed_pass_1.csv')
# Create a 2D histogram (heatmap) of court coordinates
heatmap, xedges, yedges = np.histogram2d(match_data['ball_location_x'], match_data['ball_location_y'], bins=20)

# Create extent for heatmap
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Plot the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(heatmap.T, cmap='hot', extent=extent, origin='lower')
plt.colorbar(label='Frequency')
plt.title('Court Heatmap')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```
